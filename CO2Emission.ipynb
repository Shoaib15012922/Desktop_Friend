```python
#ML Model
```

```python
data.rename(columns={'Engine Size(L)': 'Engine_Size', 'Fuel Consumption Comb (L/100 km)': 'Fuel_Cons',"Fuel Type": "Fuel_Type","CO2 Emissions(g/km)":"Emission"}, inplace=True)

```

```python
data.head()
```

```python
model = smf.glm(formula = "Emission ~ Fuel_Type * Engine_Size * Fuel_Cons", data = data)
result = model.fit()

```

```python
result.summary()
```
